```python
import numpy as np
import matplotlib.pyplot as plt

entities = {1:38, 4:14, 9:31, 16:6, 21:42, 28:84, 36:44,
                  47:26, 49:11, 51:67, 56:53, 62:19, 64:60,
                  71:91, 80:100, 87:24, 93:73, 95:75, 98:78}
```

```python
np.random.seed(42)

moves_per_game = []

for _ in range(int(1e4)): # 1 million Monte Carlo simulations
    
    pos, moves = 0, 0;
    
    while pos < 100:
        moves += 1;
        die_result = np.random.randint(1,7)
        pos += die_result;

        #if pos > 100:
        #    pos += die_result;
        #    break       

        pos = entities.get(pos,pos)
    
    moves_per_game.append(moves)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('_classic_test')

plt.hist(moves_per_game, bins=range(200), density=True, alpha=1)
plt.xlabel('Number of moves')
plt.ylabel('Relative frequency')
plt.title('Snakes & Ladders - Simulated');
```

```python
# Die probability
p = 1/6

# Transition matrix
T = np.zeros((101, 101))

for k in range(1,101):
    T[k-1, k:k+6] = p
    
T[95:100,100] += np.linspace(1/6, 5/6, 5)

print(T[99,100])

player_pos = np.zeros(101)
player_pos[0] = 1

#cl_mat = np.zeros((101, 101))
#ind = [entities.get(i, i) for i in range(101)]
#cl_mat[ind, range(101)] = 1

#mat = cl_mat @ T
```

```python
for pos_init in sorted(entities):
    
    T[:,pos_init] = 0;
    iw = np.where(T[pos_init,:] > 0)
    T[iw,entities[pos_init]] += p
    
    if pos_init > entities[pos_init]:
        T[pos_init,100] = 0;
        
v = np.zeros(101)
v[0] = 1

n, P = 0, []
cumulative_prob = 0
# Update the state vector v until the cumulative probability of winning
# is "effectively" 1
while cumulative_prob < 0.99999:
    n += 1
    v = v.dot(T)
    P.append(v[100])
    cumulative_prob += P[-1]
mode = np.argmax(P)+1
print('modal number of moves:', mode)

# Plot the probability of winning as a function of the number of moves
fig, ax = plt.subplots()
ax.plot(np.linspace(1,n,n), P, 'g-', lw=2, alpha=0.6, label='Markov')
ax.set_xlabel('Number of moves')
ax.set_ylabel('Probability of winning')

T[:,100]
```

```python
ladders = [(1,38), (4,14), (9,31), (21,42), (28,84),
           (36,44), (51,67), (71,91), (80,100)]
snakes = [(16,6), (48,26), (49,11), (56,53), (62,19),
          (64,60), (87,24), (93,73), (95,75), (98,78)]
trans = ladders + snakes

# Set up the transition matrix
T = np.zeros((101, 101))
for i in range(1,101):
    T[i-1,i:i+6] = 1/6

for (i1,i2) in trans:
    iw = np.where(T[:,i1] > 0)
    T[:,i1] = 0
    T[iw,i2] += 1/6

# House rules: you don't need to land on 100, just reach it.
T[95:100,100] += np.linspace(1/6, 5/6, 5)
for snake in snakes:
    T[snake,100] = 0
```
